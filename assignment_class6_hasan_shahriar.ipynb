# One Hot Encoding

```python
one_hot=pd.get_dummies(df,drop_first=True)
```

```python
one_hot.head()
```

# Label Encoder

```python
lbl_enc=preprocessing.LabelEncoder()
```

```python
df.apply(lbl_enc.fit_transform)
df.head()
```

# Label Encoder for loop

```python
for column in df.columns:
    if is_numeric_dtype(df[column]):
        continue
        
    else:
        df[column]=lbl_enc.fit_transform(df[column])
```

```python
df.head()
```

# Ordinal Encoding

```python
ord_enc=preprocessing.OrdinalEncoder()
```

```python
ord_enc.fit_transform(df)
df.head()
```

# Ordinal Encoding for loop

```python
for column1 in df.columns:
    if is_numeric_dtype(df[column1]):
        continue
        
    else:
        df[column1]=ord_enc.fit_transform(df[[column1]])
```
